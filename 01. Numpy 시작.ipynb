## 정렬(sort)
- 2차원 이상일 경우에는 행이나 열을 각각 따로 정렬
- axis = 0이면 각 열을 기준으로 정렬, axis = 1이면 각 행을 기준으로 정렬
- 생략하면 axis = 0
- axis = -1가 주어지면 안쪽차원이 기준이 된다.

```python
a = np.array([[4,3,5,7],
              [1,12,11,9],
              [2,15,1,14]])
a
```

```python
np.sort(a) # 여기서 생략하면 axis=1은 axis=-1과 동일하다.
```

```python
np.sort(a, axis=0)
```

```python
np.sort(a, axis=-1)
```

```python
a.sort(axis=1)
a
```

```python
# 정렬을 하지 않고 순서만 알고 싶다면 argsort 명령을 사용한다.
a = np.array([4,2,3,8])
b = np.argsort(a)
b
```

```python
a[b]
```

```python
np.sort(a)
```

## Fancy lndexing : 펜시인덱싱
- boolean 인덱싱을 이용하는 것
- 마스크 인덱싱
- 규칙이 일정하지 않지만 원하는 데이터를 추출하고자 할 때 사용한다.

```python
names = np.array(['park','park','kim','joan','lee','park','hong','park'])
names
```

```python
names.shape
```

```python
# 정규분포 : np.random.randn(), 평균 0 표준편차 1인
# np.random.rand(): 0 ~ 1 사이의 난수 발생
data = np.random.randn(8,4)
data
```

```python
data.shape
```

```python
index = (names == 'park')
data[index, :]
```

```python
# 요소가 'kim'인 행의 데이터만 추출
data[names == 'kim', :]
```

```python
data[(names=='kim') | (names =='hong'),:]
```

```python
# 0번째 열의 값이 0보다 작은 행을 추출
data[:, 0] < 0
```

```python
data[data[:, 0] < 0, :]
```

```python
# 0번째 열의 값이 0보다 작은 행의 2, 3번째 열의 값만 추출
data[data[:, 0] < 0, 2:4]
```

```python
# 0번째 열의 값이 0보다 작은 행의 2,3 번째 열의 값을 0으로 치환
data[data[:, 0] < 0, 2:4] = 0
data
```
