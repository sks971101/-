- 대구와 대전의 경우는 두 시리즈에 공통의 인덱스로 존재하지 않기 때문에 연산을 수행할 수 없어 그 결과값으로 NAN(Not a Number) 값을 가진다.
- NaN값이 float 자료형에서만 적용되므로 연산의 결과가 모두 float형으로 바뀌었다.

```python
# NaN값이 아닌 값을 구하려면 notnull메서드를 사용한다.
ds.notnull()
```

```python
ds[ds.notnull()]
```

```python
# 2010년도 자료와 2015년도 자료를 이용하여 인구 증가율(%)을 계산하시오.
res = (s-s2)/s2 * 100
res = res[res.notnull()]
res
```

### 1.3.1 데이터프레임 생성
- 여러개의 시리즈객체가 묶어서 생성된다.
- 딕셔너리를 이용(키가 열인덱스가 된다.)
- 속성 : data, index, dtype, columns

```python
data = {
    "2015": [9904312, 3448737, 2890451, 2466052],
    "2010": [9631482, 3393191, 2632035, 2431774],
    "2005": [9762546, 3512547, 2517680, 2456016],
    "2000": [9853972, 3655437, 2466338, 2473990],
    "지역": ["수도권", "경상권", "수도권", "경상권"],
    "2010-2015 증가율": [0.0283, 0.0163, 0.0982, 0.0141]
}

columns = ['지역','2015','2010','2005','2000','2010-2015 증가율']
index = ['서울','부산','인천','대구']

df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
# 데이터 값에만 접근 values 속성
# 열인덱스와 행인덱스는 각각 columns, index 속성
df.values
```

```python
df.columns
```

```python
df.index
```

```python
# 속성들에 이름을 부여할 때 name 속성
df.index.name = '도시'
df.columns.name = '특성'
df
```

```python
# 전치(transpose)
df.T
```

```python
# '2005-2010 증가율' 열 추가
df['2005-2010 증가율'] = (((df['2010']-df['2005'])/df['2005'])*100).round(2)
df
```

```python
# 두 데이터프레임에 공통의 키가 존재하는 데이터만 합치는 것 : inner join
pd.merge(df1, df2)
```

```python
# outer join, 외부조인,  키 값이 한쪽에만 있어도 표시하는 join
# left outer join, right outer join
pd.merge(df1,df2, how='outer')
```

```python
pd.merge(df1,df2, how='left')
```

```python
pd.merge(df1,df2, how='right')
```

## 1.6 그룹분석
- 특정 키를 조건으로 그룹을 설정하여 그룹에 특성을 이용하여 계산을 수행하는 기능
- groupby 메서드

**그룹연산 메서드**
- size, count : 그룹 데이터의 갯수
- mean, median, min, max
- sum, prod, std, var, quantile
- first, last

**그룹연산 메서드**
- agg, aggreate : 그룹에 동시에 적용할 함수를 리스트로 전달
- describe
- apply
- transform

```python
df = pd.DataFrame({
    'key1':['A','A','B','B','A'],
    'key2':['one','two','one','two','one'],
    'data1':[1,2,3,4,5],
    'data2':[10,20,30,40,50]
})
df
```

```python
# A그룹과 B그룹으로 구분
groups = df.groupby(df.key1)
groups
```

```python
# groups속성 : 각 그룹별 데이터의 인덱스가 저장된다.
groups.groups
```

```python
groups.sum()
```

```python
groups = df.groupby(df.key1).sum()
groups
```

```python
df['data1'].groupby(df.key1).sum()
```

```python
df.groupby(df.key1)['data1'].sum()
```

```python
df.groupby(df.key1).sum()['data1']
```

```python
df.data1.groupby([df.key1, df.key2]).sum()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(10)
```

- total_bill : 총결제금액
- tip : 팁
- sex : 결제한 사람의 성별
- smoker : 흡연/금연 여부
- day : 방문요일
- time : 시간
- size : 인원수

```python
# 결제금액대비 팁의 비율
tips['tip_pct'] = tips['tip']/tips['total_bill']
tips.head(5)
```

```python
tips.info()
```

```python
tips.describe()
```

```python
tips.groupby('sex').count()
```

```python
tips.groupby('sex').size()
```

```python
# 성별에 따른 흡연여부
tips.groupby(['sex','smoker']).size()
```

```python
# 성별별 평균 팁 비율
tips.groupby('sex')[['tip_pct']].mean()
```

```python
# 흡연여부에 따른 평균 팁 비율
tips.groupby('smoker')[['tip_pct']].mean()
```

```python
tips['day'].value_counts()
```

```python
tips['time'].value_counts()
```

```python
# 요일별로 팁 비율 분석
tips.groupby('day')[['tip_pct']].describe().T
```

```python
# 시간대별로 팁 비율 분석
tips.groupby('time')['tip_pct'].describe().T
```

```python
# 인원수별로 팁 비율 분석
tips.groupby('size')[['tip_pct']].describe().T
```

```python
# 점심일때 요일별 팁 비율 분석
tips[tips['time'] == 'Lunch'].groupby('day')[['tip_pct']].describe().T
```
