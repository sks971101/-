## Series 객체

- 구성요소 : 값(value) + 인덱스(index)
- 속성 : Series.index, Series.values, Series.dtype
- 형식 : pd.Series(data, index)

```python
# in, not in
# 사전을 바로 시리즈 객체로 변환이 가능하다.
# 사전에 키가 시리즈 인덱스가 된다.
ser
```

```python
'a' in ser
```

```python
'f' not in ser
```

```python
# 시리즈 객체를 사전취급 가능
# 사전.keys()
# 사전.values()
# 사전.items()

for key, value in ser.items():
  print("%s : %d" % (key, value))
```

```python
# 데이터사전을 이용한 데이터프레임 생성
df = pd.DataFrame({
    'weight':[80,70.4,65.5,40.9,51.2],
    'height':[170,180,155,143,154],
    'gender':['f','m','m','f','f']
})
df
```

```python
print(type(df))
print(type(df['weight']))
```

```python
df['weight']
```

```python
df[['weight','gender']] # 2개 부터 리스트로 묶어야한다.
```

```python
# height값이 160이상인 자료만 추출
df['height'] >= 160
df[df['height'] >= 160]
```

```python
# 성별이 여자인 자료만 추출
df[df['gender'] == 'f']
```

```python
df[df.gender == 'f'] # 이렇게 점을 찍을수도 있다.
```
